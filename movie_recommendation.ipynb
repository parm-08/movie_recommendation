```python
#STEP1--> Remove all punctuations and stopwords.
#STEP2-->Lemmatization
#STEP3--> Split words -Make tokens out of words.
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re  #re means regular expressions for removing punctuators.
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words=set(stopwords.words('english'))
lemmatizer=WordNetLemmatizer()
```

```python
def preprocess_text(text) : #Tags vala text
   #1 lowercase convertion
   text=str(text).lower()
   #2 removing punctuations
   text=re.sub(r'[^a-zA-Z\s]','',text)
   words=text.split()
  #3 Remove stopwords
   words=[word for word in words if word not in stop_words]
   #4 Lemmatize the words
   words=[lemmatizer.lemmatize(word) for word in words if word not in stop_words]
   return " ".join(words)
```

```python
df['tags']=df['tags'].apply(preprocess_text)
```

```python
df['tags'][1]
```

```python
#Now convert each sentence into vectors(vectorization).
# TF-IDF (Text----> Vectors)
```

```python
# Resetting the index
df=df.reset_index(drop=True)
```

```python
indices=pd.Series(df.index,index=df['title']).drop_duplicates()
indices
```

```python
#Applying vectorization
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer(max_features=50000,ngram_range=(1,2),stop_words='english')
```

```python
tfidf_matrix=tfidf.fit_transform(df['tags'])
tfidf_matrix
```

```python
#COSINE SIMILARITY
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
def recommend(title,n=10) :#n=10 means 10 movies recommendation me show hongi
  if title not in indices:
    return ["MOVIE NOT FOUND!"] #means if user types such as movie which as no similar movies return movie not found.

  idx=indices[title]
  sim_score= cosine_similarity(tfidf_matrix[idx],tfidf_matrix).flatten()
  similar_idx=sim_score.argsort()[::-1][1:n+1]
  return df['title'].iloc[similar_idx]
```

```python
recommend("Toy Story")
```

```python
recommend('Avenger')
```

```python
recommend('Avatar',5)
```
